```python
def func(x): return 0.5 * np.exp(x) + 1    # The function definition.
a, b = 0.5, 1.5    # The integral limits.
x = np.linspace(0, 2)    # The x values to plot the function.
y = func(x)    # The y values to plot the function.
Ix = np.linspace(a, b)    # The x values within the integral limits.
Iy = func(Ix)    # The y values within the integral limits.
# The list object with multiple tuple objects representing coordinates for
# the polygon to be plotted.
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
```

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', linewidth=2)     # Plots the function values as a blue line.
plt.ylim(bottom=0)    # Defines the minimum y value for the ordinate axis.
# Plots the polygon (integral area) in gray.
poly = Polygon(verts, facecolor='.7', edgecolor='.5')
ax.add_patch(poly)
# Places the integral formula in the plot.
plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$', horizontalalignment='center', fontsize=20)
# Places the axis labels.
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
# Places the x labels.
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
# Places the y labels.
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'));
```

# Static 3D Plotting

```python
strike = np.linspace(50, 150, 24)    # The ndarray object with the strike values.
```

```python
ttm = np.linspace(0.5, 2.5, 24)    # The ndarray object with the time-to-maturity values.
```

```python
strike, ttm = np.meshgrid(strike, ttm)    # The two two-dimensional ndarray objects (grids) created.
```

```python
strike[:2].round(1)    # The two two-dimensional ndarray objects (grids) created.
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm    # The dummy implied volatility values.
```

```python
iv[:5, :3]    # The dummy implied volatility values.
```

```python
# Imports the relevant 3D plotting features, which is required although
# Axes3D is not directly used.
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.gca(projection='3d')    # Sets up a canvas for 3D plotting.
# Creates the 3D plot.
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True)
ax.set_xlabel('strike')    # Sets the x-axis label.
ax.set_ylabel('time-to-maturity')    # Sets the y-axis label.
ax.set_zlabel('implied volatility')    # Sets the z-axis label.
fig.colorbar(surf, shrink=0.5, aspect=5);    # Creates a color bar.
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)    # Sets the viewing angle.
ax.scatter(strike, ttm, iv, zdir='z', s=25, c='b', marker='^')    # Creates a 3D scatter plot.
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility');
```

## Basic Plots

```python
import pandas as pd
```

```python
import cufflinks as cf    # Imports Cufflinks.
```

```python
import plotly.offline as plyo    # plotly.offline as plyo
```

```python
plyo.init_notebook_mode(connected=True)    # Turns on the notebook plotting mode.
```

```python
a = np.random.standard_normal((250, 5)).cumsum(axis=0)    # The standard normally distributed pseudo-random numbers.
```

```python
index = pd.date_range('2019-1-1',    # The start date for the DatetimeIndex object.
                      freq='B',    # The frequency (“business daily”).
                      periods=len(a))    # The number of periods needed.
```

```python
df = pd.DataFrame(100 + 5 * a,    # A linear transform of the raw data.
                  columns=list('abcde'),    # The column headers as single characters.
                  index=index)    # The DatetimeIndex object.
```

```python
df.head()    # The first five rows of data.
```

```python
plyo.iplot(    # This makes use of the offline (notebook mode) capabilities of plotly.
    # The df.iplot() method is called with parameter asFigure=True to allow
    # for local plotting and embedding.
    df.iplot(asFigure=True),
    # image='png',    # The image option provides in addition a static bitmap version of the plot.
    # The filename for the bitmap to be saved is specified (the file type extension
    # is added automatically).
    filename='ply_01'
)
```

```python
plyo.iplot(
    df[['a', 'b']].iplot(asFigure=True,
        theme='polar',    # Selects a theme (plotting style) for the plot.
        title='A Time Series Plot',    # Adds a title.
        xTitle='date',    # Adds an x-axis label.
        yTitle='value',   # Adds a y-axis label.
        mode={'a': 'markers', 'b': 'lines+markers'},    # Defines the plotting mode (line, marker, etc.) by column.
        symbol={'a': 'circle', 'b': 'diamond'},    # Defines the symbols to be used as markers by column.
        size=3.5,    # Fixes the size for all markers.
        colors={'a': 'blue', 'b': 'magenta'},    # Specifies the plotting color by column.
    ),
    # image='png',
    filename='ply_02'
)
```

```python
plyo.iplot(
    df.iplot(kind='hist',    # Specifies the plotting type.
             subplots=True,    # Requires separate subplots for every column.
             bins=15,    # Sets the bins parameter (buckets to be used = bars to be plotted).
             asFigure=True),
    # image='png',
    filename='ply_03'
)
```

## Financial Plots

```python
raw = pd.read_csv('fxcm_eur_usd_eod_data.csv', index_col=0, parse_dates=True)
# Reads the financial data from a CSV file.
```

```python
# The resulting DataFrame object consists of multiple columns and more than
# 2,800 data rows.
raw.info()
```

```python
# Selects four columns from the DataFrame object (Open-High-Low-Close, or
# OHLC).
quotes = raw[['OpenAsk', 'HighAsk', 'LowAsk', 'CloseAsk']]
quotes = quotes.iloc[-60:]    # Only a few data rows are used for the visualization.
quotes.tail()    # Returns the final five rows of the resulting data set quotes.
```

```python
qf = cf.QuantFig(
    quotes,    # The DataFrame object is passed to the QuantFig constructor.
    title='EUR/USD Exchange Rate',    # This adds a figure title.
    legend='top',    # The legend is placed at the top of the plot.
    name='EUR/USD'    # This gives the data set a name.
)
```

```python
plyo.iplot(
    qf.iplot(asFigure=True),
    # image='png',
    filename='qf_01'
)
```

```python
qf.add_bollinger_bands(periods=15,    # The number of periods for the Bollinger band.
                       boll_std=2)    # The number of standard deviations to be used for the band width.
```

```python
plyo.iplot(qf.iplot(asFigure=True),
    # image='png',
    filename='qf_02'
)
```

```python
qf.add_rsi(periods=14,    # Fixes the RSI period.
           showbands=False)    # Does not show an upper or lower band.
```

```python
plyo.iplot(
    qf.iplot(asFigure=True),
    # image='png',
    filename='qf_03'
)
```
